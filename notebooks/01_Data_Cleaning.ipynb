```python
df = df.drop_duplicates()

print("New Shape:", df.shape)
```

```python
df["Date of Admission"] = pd.to_datetime(df["Date of Admission"])
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"])
```

```python
df[["Date of Admission", "Discharge Date"]].head()
```

```python
text_columns = [
    "Name",
    "Gender",
    "Blood Type",
    "Medical Condition",
    "Doctor",
    "Hospital",
    "Insurance Provider",
    "Admission Type",
    "Medication",
    "Test Results"
]

for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
df.head()
```

```python
df["Length of Stay"] = (
    df["Discharge Date"] - df["Date of Admission"]
).dt.days

df[["Date of Admission", "Discharge Date", "Length of Stay"]].head()
```

```python
print("Minimum Length of Stay:", df["Length of Stay"].min())
print("Maximum Length of Stay:", df["Length of Stay"].max())
```

```python
df["Senior Citizen"] = df["Age"] >= 65
df["Emergency Admission"] = df["Admission Type"] == "Emergency"
df["Abnormal Test"] = df["Test Results"] == "Abnormal"
```

```python
df["Risk Score"] = (
    df["Senior Citizen"].astype(int)
    + df["Emergency Admission"].astype(int)
    + df["Abnormal Test"].astype(int)
)

df[["Age", "Admission Type", "Test Results", "Risk Score"]].head()
```

```python
df["Risk Score"] = (
    df["Senior Citizen"].astype(int)
    + df["Emergency Admission"].astype(int)
    + df["Abnormal Test"].astype(int)
)

df[["Age", "Admission Type", "Test Results", "Risk Score"]].head()
```

```python
def get_risk_category(score):
    if score == 0:
        return "Low"
    elif score == 1:
        return "Medium"
    elif score == 2:
        return "High"
    else:
        return "Critical"

df["Risk Category"] = df["Risk Score"].apply(get_risk_category)

df[["Risk Score", "Risk Category"]].head()
```

```python
df.to_csv("../data/cleaned_healthcare_dataset.csv", index=False)

print("✅ Cleaned dataset saved successfully!")
```
