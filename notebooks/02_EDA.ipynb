```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/cleaned_healthcare_dataset.csv")
```

```python
df["Date of Admission"] = pd.to_datetime(df["Date of Admission"])
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"])
```

```python
df.head()
```

```python
print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Age"], bins=20)

plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Number of Patients")

plt.show()
```

```python
gender_counts = df["Gender"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Gender Distribution")

plt.show()
```

```python
condition_counts = df["Medical Condition"].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(condition_counts.index, condition_counts.values)

plt.xticks(rotation=45)
plt.title("Medical Conditions")
plt.xlabel("Condition")
plt.ylabel("Count")

plt.show()
```

```python
admission_counts = df["Admission Type"].value_counts()

plt.figure(figsize=(7, 5))
plt.bar(admission_counts.index, admission_counts.values)

plt.title("Admission Types")
plt.xlabel("Admission Type")
plt.ylabel("Patients")

plt.show()
```

```python
test_counts = df["Test Results"].value_counts()

plt.figure(figsize=(7, 5))
plt.bar(test_counts.index, test_counts.values)

plt.title("Test Results")
plt.xlabel("Result")
plt.ylabel("Count")

plt.show()
```

```python
blood_counts = df["Blood Type"].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(blood_counts.index, blood_counts.values)

plt.title("Blood Group Distribution")
plt.xlabel("Blood Type")
plt.ylabel("Patients")

plt.show()
```

```python
insurance_counts = df["Insurance Provider"].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(insurance_counts.index, insurance_counts.values)

plt.xticks(rotation=45)

plt.title("Insurance Providers")
plt.xlabel("Provider")
plt.ylabel("Patients")

plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Billing Amount"], bins=30)

plt.title("Billing Amount Distribution")
plt.xlabel("Billing Amount")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Length of Stay"], bins=20)

plt.title("Length of Stay")
plt.xlabel("Days")
plt.ylabel("Patients")

plt.show()
```

```python
risk_counts = df["Risk Category"].value_counts()

plt.figure(figsize=(7, 5))
plt.bar(risk_counts.index, risk_counts.values)

plt.title("Risk Categories")
plt.xlabel("Risk Level")
plt.ylabel("Patients")

plt.show()
```

```python
billing = df.groupby("Admission Type")["Billing Amount"].mean()

plt.figure(figsize=(7, 5))
plt.bar(billing.index, billing.values)

plt.title("Average Billing by Admission Type")
plt.xlabel("Admission Type")
plt.ylabel("Average Billing")

plt.show()
```

```python
print("Average Age:", round(df["Age"].mean(), 2))
print("Average Billing:", round(df["Billing Amount"].mean(), 2))
print("Average Length of Stay:", round(df["Length of Stay"].mean(), 2))
print("Total Patients:", len(df))
```
